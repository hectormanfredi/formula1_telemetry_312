# Capture data from api OpenF1
* Take all the data from the site and storage in parquet files or some database
* create dashsboard with streamlit

```python
import json
import pandas as pd
from urllib.request import urlopen


response = urlopen('https://api.openf1.org/v1/sessions?date_start>=2021-09-01&date_end<=2024-11-10')
data = json.loads(response.read().decode('utf-8'))

df = pd.DataFrame(data)
```
